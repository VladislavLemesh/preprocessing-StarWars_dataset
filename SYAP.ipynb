#Задание 2. Нужно переименовать столбцы. Столбцы 4-9 означают смотрел ли человек 1 часть, 2 и т.д. Столбцы 10-15 оценка каждой части отдельно

```python
df.columns = ('RespondentID',
       'seen any of the 6 films',
       'fan of the Star Wars',
       'seen 1 Star Wars films',
       'seen 2 Star Wars films', 'seen 3 Star Wars films', 'seen 4 Star Wars films', 'seen 5 Star Wars films', 'seen 6 Star Wars films',
       'rank the 1 Star Wars films',
       'rank the 2 Star Wars films', 'rank the 3 Star Wars films', 'rank the 4 Star Wars films', 'rank the 5 Star Wars films',
       'rank the 6 Star Wars films',
       'fan of the Star Trek',
       'Gender', 'Age', 'Education', 'Location (Census Region)')
```

```python
df
```

```python
#Заполнение столбцов с отсутствующими значениями с вопросом "смотрел ли вы фильм" т.к отсутсвие значения означает что не смотрел фильм
feature_seen_film_list = ['seen 1 Star Wars films',
       'seen 2 Star Wars films', 
       'seen 3 Star Wars films', 
       'seen 4 Star Wars films', 
       'seen 5 Star Wars films', 
       'seen 6 Star Wars films']
df[feature_seen_film_list] = df[feature_seen_film_list].fillna('No')
```

```python
df = df.dropna()
```

# Задание 3. Столбцы с рейтингом перевести во float. Далее группируем данные по полу и получаем среднею оценку по каждому фильму. После чего получить список людей где мужчины оценили 4-6 часть выше среднего, а 1-3 часть ниже среднего. И женщины которые оценили все фильмы ниже среднего

```python
feature_rank_list = ['rank the 1 Star Wars films',
       'rank the 2 Star Wars films', 
       'rank the 3 Star Wars films', 
       'rank the 4 Star Wars films', 
       'rank the 5 Star Wars films',
       'rank the 6 Star Wars films']
df.loc[:, feature_rank_list] = df[feature_rank_list].astype(float)
```

```python
#df.isna().mean()
df
```

```python
df['seen 1 Star Wars films'][df['seen 1 Star Wars films'] != 'No'] = 'Yes'
df['seen 2 Star Wars films'][df['seen 2 Star Wars films'] != 'No'] = 'Yes'
df['seen 3 Star Wars films'][df['seen 3 Star Wars films'] != 'No'] = 'Yes'
df['seen 4 Star Wars films'][df['seen 4 Star Wars films'] != 'No'] = 'Yes'
df['seen 5 Star Wars films'][df['seen 5 Star Wars films'] != 'No'] = 'Yes'
df['seen 6 Star Wars films'][df['seen 6 Star Wars films'] != 'No'] = 'Yes'
df
```

```python
first_request = df[['rank the 1 Star Wars films',
       'rank the 2 Star Wars films', 
       'rank the 3 Star Wars films', 
       'rank the 4 Star Wars films', 
       'rank the 5 Star Wars films',
       'rank the 6 Star Wars films', 'Gender']].groupby('Gender').mean()

first_request
```

```python
mean_rank_male = first_request[feature_rank_list].iloc[1]
```

```python
mean_rank_women = first_request[feature_rank_list].iloc[0]
```

```python
second_request = df[((df['Gender'] == 'Male') 
& (df['rank the 4 Star Wars films'] > mean_rank_male[3]) 
& (df['rank the 5 Star Wars films'] > mean_rank_male[4]) 
& (df['rank the 6 Star Wars films'] > mean_rank_male[5]) 
& (df['rank the 1 Star Wars films'] < mean_rank_male[0]) 
& (df['rank the 2 Star Wars films'] < mean_rank_male[1]) 
& (df['rank the 3 Star Wars films'] < mean_rank_male[2])) 
| ((df['Gender'] == 'Female') 
& (df['rank the 1 Star Wars films'] < mean_rank_women[0]) 
& (df['rank the 2 Star Wars films'] < mean_rank_women[1]) 
& (df['rank the 3 Star Wars films'] < mean_rank_women[2]) 
& (df['rank the 4 Star Wars films'] < mean_rank_women[3]) 
& (df['rank the 5 Star Wars films'] < mean_rank_women[4]) 
& (df['rank the 6 Star Wars films'] < mean_rank_women[5]))]
```
